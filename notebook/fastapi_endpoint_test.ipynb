```python
from supabase import create_client
import os
from dotenv import load_dotenv
from config import CLEAN_DATA_NAME, FEATURE_STORE_NAME

load_dotenv()

SUPABASE_URL = os.getenv("SUPABASE_URL")
SUPABASE_KEY = os.getenv("SUPABASE_PUBLIC_KEY")

supabase = create_client(SUPABASE_URL, SUPABASE_KEY)

response = supabase.table(CLEAN_DATA_NAME).select("category").execute()

unique_categories = list(set(
    row["category"] for row in response.data
))

print(unique_categories)
```
